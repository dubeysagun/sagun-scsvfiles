```python
# fetching one column
print(file['Flight Distance'],"\n\n\n")
#fetching multi columns
print(file[['Flight Distance','Seat comfort']],"\n\n\n")
#fetching 1st row
print(file.iloc[0],"\n\n\n")
#fetching 2nd and 3rd row
print(file.iloc[1:3],"\n\n\n")
#fetching particular rows
print(file.iloc[[1,3,6]])
```

```python
# filtering dataframe on a condition
#printing details of customers of business class
mask=file['Class']=='Business'
print(file[mask],"\n\n\n")
#printing no. of customers in Business class
print(file[mask].shape[0],"\n\n\n")
```

```python
# finding different categories of a given column no. of rows having each category
print(file['Class'].value_counts(),"\n\n\n")
#plotting no. of customers in different class using bar graph
file['Class'].value_counts().plot(kind='bar')
```

```python
#plotting no. of customers in different class using pie chart in pandas 
file['Class'].value_counts().plot(kind='pie')
```

```python
#plotting no. of customers in different class in the form of pie chart using matplotlib....

plt.pie(file['Class'].value_counts(),labels=(file['Class'].value_counts()).index,autopct='%1.0f%%')
plt.show()
```
